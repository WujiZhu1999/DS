# Part 1 Linear Regression from Scratch

The idea is to solve:

Based on the error function:

We want to calculate the gradient (of the loss function) which is the direction of greatest increase loss.

By getting derivative we know that

Then we update the weight to learn model:


```python
## Prepare data for linear regression
## x: feature, y:target
import numpy as np
np.random.seed(12345)
x = np.array([1, 2, 1.2, 1.5])
y = x * 2 + np.random.randn(4)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(x, y)
plt.xlabel("house area")
plt.ylabel("house price")
```

```python
# define loss function
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# define gradient calculation function
def grad(y_true, w, x):
    return np.mean(2 * (y_true - w*x) * -x)

# gradient descent step
def grad_decent(y_true, w, x, learning_rate):
    w = w - learning_rate * grad(y_true, w, x)
    return w

w = np.random.rand()
learning_rate = 0.001
num_iteration = 2000
loss = []
for step in range(num_iteration):
    w = grad_decent(y, w, x, learning_rate)
    loss_step = mse(y, w*x)
    loss.append(loss_step)
```

```python
plt.plot(loss)
plt.xlabel("#iteration")
plt.ylabel("mse_loss")
```

```python
plt.scatter(x, y, label="true")
plt.scatter(x, w*x, label="predict", marker="^")
plt.legend()
```

# Part 2 Common Loss Functions

```python
# MAE (as formula above)
y_pred = w * x
from sklearn.metrics import mean_absolute_error as mae
mae(y, y_pred)
```

```python
# MSE as shown above
from sklearn.metrics import mean_squared_error as mse
mse(y, y_pred)
```


Closer to 1 ==> better

```python
# R^2
from sklearn.metrics import r2_score as r2
r2(y, y_pred)
```

---

2. Understand it by how it affect gradient descent.

L1：是绝对值之和。当w大于0时，梯度始终为正常数，更新的参数w变小；当w小于0时，梯度始终为负常数，更新的参数w变大；所以，L1正则化容易使参数变为0，即特征稀疏化。

L2：是平方和。当w趋向于0时，参数减小的非常缓慢，因此L2正则化使参数减小到很小的范围，但不为0。

---

3. Where it is useful?

L1：使模型中尽可能多的参数值为0，因此适用于：模型剪枝，模型压缩，特征选择。是一种从改变模型结构的角度（减少模型参数的数量）解决过拟合的方式。

L2：使模型中的所有参数值尽可能小，使得模型尽量不依赖于某几个特殊的特征，而是使每个特征都得到尽量均衡的权重，因此适用于解决普通的过拟合问题，即从参数分布（让分布尽可能的均匀）的角度解决过拟合的问题，这也是常用的解决过拟合的方式。
